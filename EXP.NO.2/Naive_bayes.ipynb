```python

from sklearn.datasets import load_wine
data = load_wine()


```

```python
x=data.data
y=data.target
```

```python
from sklearn.naive_bayes import GaussianNB
clf_1=GaussianNB()
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size =.8)
ML_1=clf_1.fit(x_train, y_train)
from sklearn.naive_bayes import MultinomialNB
clf_2=MultinomialNB()
ML_2=clf_2.fit(x_train,y_train)
from sklearn.naive_bayes import BernoulliNB
clf_3=BernoulliNB(binarize=2)
ML_3=clf_3.fit(x_train,y_train)
y_pred1=clf_1.predict(x_test)
y_pred2=clf_2.predict(x_test)
y_pred3=clf_3.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_1=accuracy_score(y_test,y_pred1)
print("Gaussian :",accuracy_1*100)
accuracy_2=accuracy_score(y_test,y_pred2)
print("Multinomial : ",accuracy_2*100)
accuracy_3=accuracy_score(y_test,y_pred3)
print("Bernoulli : ",accuracy_3*100)
```

```python
import pandas as pd
file_2=pd.read_csv("/content/HepatitisCdata (2).csv")
x=file_2[["Age","Sex","ALB","ALP","ALT","AST","BIL","CHE","CHOL","CREA","GGT","PROT"]]
y=file_2["Category"]
x
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
x["Sex"]=lb.fit_transform(x["Sex"])
x=x.fillna(0)
x
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8)
```

```python
from sklearn.naive_bayes import GaussianNB
clf_1=GaussianNB()
clf_1=clf_1.fit(xtrain,ytrain)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf_2=MultinomialNB()
ml_2=clf_2.fit(xtrain,ytrain)
```

```python
from sklearn.naive_bayes import BernoulliNB
clf_3=BernoulliNB(binarize=3)
ml_3=clf_3.fit(xtrain,ytrain)
```

```python
res_1=ml_1.predict(xtest)
res_2=ml_2.predict(xtest)
res_3=ml_3.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_1=accuracy_score(ytest,res_1)
print("Gaussian :",accuracy_1*100)
accuracy_2=accuracy_score(ytest,res_2)
print("Multinomial : ",accuracy_2*100)
accuracy_3=accuracy_score(ytest,res_3)
print("Bernoulli : ",accuracy_3*100)
```
